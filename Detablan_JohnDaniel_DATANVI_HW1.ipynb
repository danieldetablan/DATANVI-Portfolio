```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec # Gridscpec allows to make a custom layout in the plot: https://matplotlib.org/stable/api/_as_gen/matplotlib.gridspec.GridSpec.html

data = pd.read_csv("tips.csv")
```

```python
# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'], c=data['size'],
            s=data['total_bill'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.colorbar()

plt.show()
```

**Bokeh**

Preliminaries

```python
from bokeh.io import output_file, show
from bokeh.layouts import row, column
from bokeh.plotting import figure
from bokeh.palettes import magma, inferno, plasma
import pandas as pd

output_file("layout.html")

data = pd.read_csv("tips.csv")
```

1. Scatter Plot


```python
# instantiating the figure object
graph = figure(title = "Bokeh Scatter Graph")

color = magma(256)

# plotting the graph
graph.scatter(data['total_bill'], data['tip'], color=color)

# displaying the model
show(graph)
```

2. Line Plot

```python
# instantiating the figure object
graph = figure(title = "Bokeh Bar Chart")

# Count of each unique value of
# tip column
df = data['tip'].value_counts()

# plotting the graph
graph.line(df, data['tip'])

# displaying the model
show(graph)
```

3. Bar Chart


```python
# instantiating the figure object
graph = figure(title = "Bokeh Bar Chart")

# plotting the graph
graph.vbar(data['total_bill'], top=data['tip'])

# displaying the model
show(graph)
```

4. Interactive Data Visualization

```python
# instantiating the figure object
graph = figure(title = "Bokeh Bar Chart")

# plotting the graph
graph.vbar(data['total_bill'], top=data['tip'],
           legend_label = "Bill VS Tips", color='green')

graph.vbar(data['tip'], top=data['size'],
           legend_label = "Tips VS Size", color='red')

graph.legend.click_policy = "hide"

# displaying the model
show(graph)
```

5. Adding Widgets

```python
# DO NOT RUN THIS CELL!

from bokeh.io import show
from bokeh.models import Button, CheckboxGroup, RadioGroup, CustomJS

button = Button(label="GFG")

button.js_on_click(CustomJS(
	code="console.log('button: click!', this.toString())"))

# Labels for checkbox and radio
# buttons
L = ["First", "Second", "Third"]

# the active parameter sets checks the selected value
# by default
checkbox_group = CheckboxGroup(labels=L, active=[0, 2])

checkbox_group.js_on_click(CustomJS(code="""
	console.log('checkbox_group: active=' + this.active, this.toString())
"""))

# the active parameter sets checks the selected value
# by default
radio_group = RadioGroup(labels=L, active=1)

radio_group.js_on_click(CustomJS(code="""
	console.log('radio_group: active=' + this.active, this.toString())
"""))

show(button)
show(checkbox_group)
show(radio_group)
```

6. Sliders

```python
# DO NOT RUN THIS CELL!


from bokeh.io import show
from bokeh.models import CustomJS, Slider

slider = Slider(start=1, end=20, value=1, step=2, title="Slider")

slider.js_on_change("value", CustomJS(code="""
    console.log('slider: value=' + this.value, this.toString())
"""))

show(slider)
```

Preliminaries

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

data = pd.read_csv("tips.csv")
```

1. Scatter Plot

```python
# plotting the scatter chart
fig = px.scatter(data, x="day", y="tip", color='sex')

# showing the plot
fig.show()
```

2. Line Chart

```python
# plotting the scatter chart
fig = px.line(data, y='tip', color='sex')

# showing the plot
fig.show()
```

3. Bar Chart

```python
# plotting the scatter chart
fig = px.bar(data, x='day', y='tip', color='sex')

# showing the plot
fig.show()
```

4. Histogram

```python
# plotting the scatter chart
fig = px.histogram(data, x='total_bill', color='sex')

# showing the plot
fig.show()
```

5. Adding Interaction

```python
# Making a Scatter Plot as Initial Graph
plot =go.Figure(data=[go.Scatter(
    x=data['day'],
    y=data['tip'],
    mode='markers',)
])

# Add dropdown
plot.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ), # Dropdown for Scatter Plot
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                ) #Dropdown for Bar Chart
            ]),
            direction="down",
        ),
    ]
)

plot.show()
```

6. Adding Buttons

```python
# Making a Scatter Plot as Initial Graph
plot = go.Figure(data=[go.Scatter(
    x=data['day'],
    y=data['tip'],
    mode='markers',)
])

# Add dropdown
plot.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="left",
            buttons=list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ), # Button for Scatter Plot
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                ) # Button for Bar Chart
            ]),
        ),
    ]
)

plot.show()
```

6. Creating Sliders and Selectors

```python
# Making a Scatter Plot as Initial Graph
plot = go.Figure(data=[go.Scatter(
    y=data['tip'],
    mode='lines',)
])

plot.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                    step="day",
                    stepmode="backward"),
            ])
        ), # Slider for Scatter Plot
        rangeslider=dict(
            visible=True
        ), # Shows the Slider
    )
)

plot.show()
```

# **Section 2**

Before

```python
# Create a figure
fig = plt.figure(figsize=(15, 10))
fig.suptitle("Before", fontsize=20)

# Creat a Gridspec Object for the desired layout
gs = gridspec.GridSpec(2, 2, height_ratios=[1, 1])

# Add subplots
sub1 = fig.add_subplot(gs[0, 0])
sub2 = fig.add_subplot(gs[0, 1])
sub3 = fig.add_subplot(gs[1, :])

# Plotting the data
sc3 = sub1.scatter(data["day"], data["tip"], c=data['size'], s=data['total_bill'])
sub1.set_title("Scatter Plot")
sub1.set_xlabel("Day")
sub1.set_ylabel("Tip")

sns.histplot(x='total_bill', ax=sub2, hue='sex', kde=True, data=data)
sub2.set_title("Histogram")
sub2.set_xlabel("Total Bill")

sns.lineplot(ax=sub3, data=data.drop(['total_bill'], axis=1))
sub3.set_title("Line Plot")
sub3.set_xlabel("Day")
sub3.set_ylabel("Tip")

plt.colorbar(sc3, ax=sub1)
plt.tight_layout()
plt.show()
```

After

```python
# Create a figure
fig = plt.figure(figsize=(15, 10))
fig.suptitle("After", fontsize=20)

# Creat a Gridspec Object for the desired layout
gs = gridspec.GridSpec(2, 2, height_ratios=[1, 1])

# Add subplots
sub1 = fig.add_subplot(gs[0, 0])
sub2 = fig.add_subplot(gs[0, 1])
sub3 = fig.add_subplot(gs[1, :])

# Plotting the data
cmap1 = plt.cm.get_cmap('Blues')
sc3 = sub1.scatter(data["day"], data["tip"], c=data['size'], s=data['total_bill'], marker='s', cmap=cmap1)
sub1.set_title("Scatter Plot")
sub1.set_xlabel("Day")
sub1.set_ylabel("Tip")

sns.histplot(x='total_bill', ax=sub2, hue='sex', palette='Set2', kde=True, data=data)
sub2.set_title("Histogram")
sub2.set_xlabel("Total Bill")

sns.lineplot(ax=sub3, data=data.drop(['total_bill'], axis=1), palette='coolwarm', marker='s')
sub3.set_title("Line Plot")
sub3.set_xlabel("Day")
sub3.set_ylabel("Tip")

plt.colorbar(sc3, ax=sub1)
plt.tight_layout()
plt.show()
```
